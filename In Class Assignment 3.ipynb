```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Define Start/End dates, and ticker names
start = datetime.datetime(2022, 1, 1)
end = datetime.datetime(2022, 2, 15)
```

```python
# read th stock information for Amazon (Tick: AMZN) using Panda's reader
AMZN = data.DataReader("AMZN", 'yahoo', start, end)
```

```python
# create new dataframe but now you have a default name of 0-252 and date has become a column 
myAMZN = AMZN.reset_index()
```

```python
# plot a time series (line chart) of the 'volume' from January 1, 2022 to today
myAMZN.plot.line(x='Date',y='Volume', xlabel='Date', ylabel='AMZN', figsize=(12,3),lw=1)
```
